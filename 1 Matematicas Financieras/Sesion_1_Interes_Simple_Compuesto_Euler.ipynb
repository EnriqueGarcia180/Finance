```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Interes Simple vs Interes Compuesto

## Interes Simple


$$
\text{M}=C\left( 1+i*n \right)   \text{; Monto Acumulado}
$$

```python
# Definimos una funcion que calcule el monto acumulado bajo un interes simple

# n - periodos
# p - frecuencia 
def int_simp(capital, tasa, periodos, frecuencia):
    monto_acumulado = capital * (1 + ((tasa/frecuencia) * periodos))
    return monto_acumulado # FV
```

```python
int_simp(1000, 0.1, 20, 12)
```

## Interes Compuesto

$$
\text{M}=C \left (1+\frac{i^{(p)}}{p} \right)^{n}    \text{; Monto Acumulado}
$$

## Tasa Equivalente

$$
i^{(p)}=\left( {(1+i)}^\frac{1}{p} - 1 \right)p
$$

```python
# Definimos una function que calcule el monto acumulado bajo un interes compuesto capitalizable n periodos

# n - periodos
# p - frecuencia 
def int_comp(capital, tasa, periodos, frecuencia):
    tasa_eq = (((1 + tasa) ** (1 / frecuencia)) - 1) * frecuencia
    monto_acumulado = capital * ((1 + (tasa_eq / frecuencia)) ** periodos)
    return monto_acumulado # FV
```

```python
int_comp(1000, 0.1, 20, 12) # $1000 de capital, a una tasa del 10% durante 6 periodos capitalizables mensualmente
```

## Grafico de Interes Simple vs Compuesto

```python
n = 25
# n: periodos en los que se capitalizara el Interes.
# es lo que vamos a ir variando para encontrar el punto 
# en el que es mejor el Interes compuesto que el simple.

capital = 1000
tasa = 1  # 100% Para que se alcance a ver en la grafica.
frecuencia = 12  # Frecuencia (p) de Capitalizacion. 12 = mensual

# Listas a llenar con el capital (FV) con el Interes calculado simple y compuesto 
# en diferentes periodos.
lista_simple = []
lista_compuesto = []

for periodos in range(0,n):
    calculo_simple = int_simp(capital, tasa, periodos, frecuencia) # Valor del capital con intere simple en x periodos
    calculo_compuesto = int_comp(capital, tasa, periodos, frecuencia)# Valor del capital con interes compuesto en x periodos
    lista_simple.append(calculo_simple)
    lista_compuesto.append(calculo_compuesto)

print(lista_simple)
print(lista_compuesto)
```

```python
# Vamos a acomodar las listas en una table (Data Frame)
mi_tabla = pd.DataFrame()
```

```python
mi_tabla['Interes Simple'] = lista_simple
mi_tabla['Interes Compuesta'] = lista_compuesto
```

```python
display(mi_tabla)
```

$$
\left\{ \text{si (p) frecuencia < 1 año: Interes Simple > Interes Compuesto} \right\}
$$
$$
\left\{ \text{si (p) frecuencia = 1 año: Interes Simple = Interes Compuesto} \right\}
$$
$$
\left\{ \text{si (p) frecuencia > 1 año: Interes Simple < Interes Compuesto} \right\}
$$

```python
# Vamos a crear un grafico de estos datos

plt.figure(figsize=(10,6), edgecolor='black', facecolor="#99CCFF")
plt.title('Interes Simple vs Interes Compuesto')
plt.xlabel("Tiempo (meses)")
plt.ylabel("Monto Acumulado")
plt.plot(mi_tabla)
plt.legend(['Interes Simple', 'Interes Compuesto'])
plt.axvline(x = 12, color = 'red', linestyle = '--')
plt.annotate(text = 't = 12 meses', xy = (12.5, 3000), color = 'red')
plt.show()
```

# Calculo del numero e (euler)

El limite al que puedo particionar el año para aumentar la frecuencia de pagos de intereses

En finanzas
$$
\left( 1+\frac{i^{(p)}}{p} \right)^{p}
$$

Relacion Matematica
$$
\lim_{x \to \infty } \left( 1+\frac{{1}}{x} \right)^{x}=e
$$

```python
# Con esta funcion determinamos el interes compuesto al final de un año

def int_comp2(capital, tasa, frecuencia):
    calculo = capital * (1 + tasa / frecuencia) ** frecuencia
    return calculo
```

```python
print('1 Peso bajo Interes Compuesto del 100% capitalizable 1 vez al año (anual) genera:',int_comp2(1,1,1))
print('1 Peso bajo Interes Compuesto del 100% capitalizable 2 veces al año (semestral) genera:',int_comp2(1,1,2))
print('1 Peso bajo Interes Compuesto del 100% capitalizable 3 veces al año (cuatrimestral) genera:',int_comp2(1,1,3))
print('1 Peso bajo Interes Compuesto del 100% capitalizable 4 veces al año (trimestral) genera:',int_comp2(1,1,4))
# ...
print('1 Peso bajo Interes Compuesto del 100% capitalizable 24 veces al año (quincenal) genera:',int_comp2(1,1,24))
# ...
print('1 Peso bajo Interes Compuesto del 100% capitalizable 52 veces al año (semanal) genera:',int_comp2(1,1,52))
#...
print('1 Peso bajo Interes Compuesto del 100% capitalizable 365 veces al año (diario) genera:',int_comp2(1,1,365))
#...
print('1 Peso bajo Interes Compuesto del 100% capitalizable 1000 veces al año genera:',int_comp2(1,1,1000))
#...
print('1 Peso bajo Interes Compuesto del 100% capitalizable 10000000 veces al año genera:',int_comp2(1,1,10000000))
print('Numero de euler e:',np.exp(1))
```

```python
# Creamos iteraciones con esta funcion para visualizar el limite, que es el numero e

# Limite cuando n -> infinito
n = 100
lista = []

for x in range(1,n):
    calculo = int_comp2(1,1,x) # $1 peso al 100% de tasa de interes capitalizable cada x periodos
    lista.append(calculo)

# Lo guardamos en un Data Frame
tabla = pd.DataFrame()

tabla['Iteraciones'] = lista

print(tabla.head())
print(tabla.tail())
```

```python
# Creamos un grafico con estas iteraciones

plt.figure(figsize=(10,6), edgecolor='black', facecolor="#99CCFF")
plt.title('Aproximacion del numero de Euler')
plt.xlabel('Particiones del año')
plt.plot(tabla)
plt.legend(['Monto Acumulado de $1 Peso bajo Interes Compuesto del 100% capitalizable n veces al año'])
plt.axhline(y = np.exp(1), color = 'r', linestyle = "--")
plt.annotate(text = 'Numero de Euler (e)', xy = (0,2.69), color = 'red')
plt.show()
```

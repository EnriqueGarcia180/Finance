### AMORTIZACIÓN GRADUAL (Frances)

```python
import numpy_financial as npf
import pandas as pd

# Ejercicio 22.
# Una persona consigue un crédito por $35,000 a pagar en 8 mensualidades 
# con una tasa de interés del 12.6% anual capitalizable por mes. 
# Elaborar una tabla de amortización.

PV = 35000
r = 0.126
n = 8

RENTAS = []
INTERESES = []
AMORTIZACIÓN = []
SALDO = []

# Periodo 0
RENTAS.append(0)
INTERESES.append(0)
AMORTIZACIÓN.append(0)
SALDO.append(f'${PV:,.2f}')

saldo = PV

# Periodos 1 al 8
for mes in range(1, n+1):

    # Usando libreria numpy_financial
    pmt = npf.pmt(r/12,n,-PV)           # Renta o abono (ver como estos seran iguales)
    ipmt = npf.ipmt(r/12, mes, n, -PV)  # Intereses
    ppmt = npf.ppmt(r/12,mes, n, -PV)   # Amortización
    saldo = saldo - ppmt                # Saldo Insoluto

    pmt = f'$ {pmt:,.2f}'
    ipmt = f'$ {ipmt:,.2f}'
    ppmt = f'$ {ppmt:,.2f}'
    saldo_2 = f'$ {saldo:,.2f}' # creamos "saldo_2" porque "saldo" lo uso en la siguiente iteracion como valor numerico, entonces saldo_2 con formato ya puede ir a la lista de saldo abajo.

    RENTAS.append(pmt)
    INTERESES.append(ipmt)
    AMORTIZACIÓN.append(ppmt)
    SALDO.append(saldo_2)


Tabla_amortizacion = pd.DataFrame()

Tabla_amortizacion['RENTAS'] = RENTAS
Tabla_amortizacion['INTERESES'] = INTERESES
Tabla_amortizacion['AMORTIZACIÓN'] = AMORTIZACIÓN
Tabla_amortizacion['SALDO_INSOLUTO'] = SALDO

Tabla_amortizacion
```

### AMORTIZACIÓN CONSTANTE (Aleman)

```python
import numpy_financial as npf
import pandas as pd

# Ejercicio 23. 
# Con un sistema de amortización constante, tasa de interés del 13.2%
# nominal mensual y plazo de 2 años, obtenga los primeros 2 pagos mensuales y 
# el último para amortizar un crédito de $96,000

PV = 96000
r = 0.132
n = 24

RENTAS = []
INTERESES = []
AMORTIZACIÓN = []
SALDO= []

RENTAS.append(0)
INTERESES.append(0)
AMORTIZACIÓN.append(0)
SALDO.append(f'${PV:,.2f}')

saldo = PV
amortizacion_constante = PV / n # Amortizacion Constante

for mes in range(1, n+1):

    ppmt = amortizacion_constante   # Amortización
    ipmt = saldo * r/12             # Intereses
    pmt = ppmt + ipmt               # Renta o abono
    saldo = saldo - ppmt            # Saldo Insoluto

    pmt = f'$ {pmt:,.2f}'
    ipmt = f'$ {ipmt:,.2f}'
    ppmt = f'$ {ppmt:,.2f}'
    saldo_2 = f'$ {saldo:,.2f}'

    RENTAS.append(pmt)
    INTERESES.append(ipmt)
    AMORTIZACIÓN.append(ppmt)
    SALDO.append(saldo_2)


Tabla_amortizacion = pd.DataFrame()

Tabla_amortizacion['RENTAS'] = RENTAS
Tabla_amortizacion['INTERESES'] = INTERESES
Tabla_amortizacion['AMORTIZACIÓN'] = AMORTIZACIÓN
Tabla_amortizacion['SALDO_INSOLUTO'] = SALDO

Tabla_amortizacion
```

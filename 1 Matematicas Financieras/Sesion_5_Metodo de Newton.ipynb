## Método de Newton para estimación de tasas de interés ***i***

```python
# definimos la función f(i)
def f_i(R,n,i,VP):
    v_n = (1 + i) ** -n
    anualidad = (1 - v_n) / i
    calculo = R * anualidad - VP
    return calculo  # Devuelve el valor de f(i)

# definimos la derivdada f'(i)
def derivada_f_i(R,n,i):
    numerador = R * i * n * (1 + i) ** (-n-1) + R * (1 + i)**(-n) - R
    denominador = i ** 2
    return numerador / denominador  # Devuelve el valor de f'(i)

# Ejercicio 15:
#       Una persona compra un refrigerador en 5,000, paga un enganche del 
#       20% sobre el valor del aparato y el resto en 12 depósitos al ﬁnal de cada 
#       periodo usando una tasa de interés efectiva del 2%.
#       Determine el monto de cada depósito."
#       Solucion:  R = $378.32, el monto de los pagos cada periodo
# AHORA, CALCULAR LA TASA DE INTERES (i) CON LOS VALORES DEL EJERCICIO 15.

R = 378.32 # Pagos en cada periodo
n = 12
VP = 4000
# i_0 = ??  
    # Es la solucion inicial propuesta.
    # ver (pag.213) de [Finan.Study.Manual.FM.pdf] donde:
    # i_0 = 2*(n-k) / (k*(n+1))
    # n = numero de periodos
    # k = anualidad durante n periodos: k = a_20 = VP/R
    # en este caso:
    # k = 4000/378.32 = 10.5730
    # i_0 = 2*(12-10.5730)/10.5730*(12+1) = 0.020764 = 2.0764%
k = VP / R
i_0 = 2*(n-k) / (k*(n+1))
print(f"i_0 = {round(i_0*100, 10)} %")

iteraciones = [] # Iteraciones del metodo de Newton. Crear una lista vacia.
iteraciones.append(round(i_0 * 100, 10))

# primer iteracion con el metodo de Newton para obtener (i)
i = i_0 -  f_i(R, n, i_0, VP) / derivada_f_i(R,n,i_0)
iteraciones.append(round(i * 100, 10))
print(f"i_1 = {round(i * 100, 10)} %")

# siguientes Iteraciones 
for x in range(1,20):
    i_x = i
    i = i_x - f_i(R, n ,i_x, VP) / derivada_f_i(R,n,i_x)
    iteraciones.append(round(i * 100, 10))
    print(f"i_{x+1} = {round(iteraciones[x+1], 10)} %")
    diff = i - i_x
    if (diff < 0.0000001):     # aqui la tasa de interes esta en decimales no en porcentaje por eso la resolucion tan baja
        break
    
print(iteraciones)          # Imprimir la lista de las iteraciones, ver que a la 3er iteracion encontro el valor.



###################
# TO-DO's, 
# 1. Implementar las graficas 
###################



```

## üìä M√©tricas de Evaluaci√≥n de una Inversi√≥n

Este bloque analiza 4 m√©tricas clave para evaluar el desempe√±o de una acci√≥n, fondo o portafolio de inversi√≥n.


### 1Ô∏è‚É£ Rendimiento Diario

> Es la variaci√≥n porcentual entre el precio de cierre de hoy y el de ayer.

**F√≥rmula:**

$$ R_d = \frac{P_t - P_{t-1}}{P_{t-1}} $$

**C√°lculo en Python:**

```python
df["Rendimiento Diario"] = df["Close"].pct_change()
df.dropna(inplace=True)  # Elimina filas con NaN
```


### 2Ô∏è‚É£ Rendimiento Acumulado

Mide el crecimiento total de la inversi√≥n desde el inicio hasta el final del per√≠odo.

F√≥rmula:

$$ R_{acum} = \prod_{t=1}^{T} (1 + R_d) - 1 $$

C√°lculo en Python:
```python
df["Rendimiento Acumulado"] = (1 + df["Rendimiento Diario"]).cumprod() - 1
```

‚∏ª

3Ô∏è‚É£ CAGR (Tasa de Crecimiento Anual Compuesta)

Muestra el rendimiento anual promedio constante, ajustado al tiempo.

F√≥rmula:

$$
\text{CAGR} = \left( \frac{V_f}{V_i} \right)^{\frac{1}{n}} - 1
$$

donde:
	‚Ä¢	( V_f ): Valor final
	‚Ä¢	( V_i ): Valor inicial
	‚Ä¢	( n ): A√±os entre inicio y fin del periodo

C√°lculo en Python:
```python
inicio = df["Close"].iloc[0]
final = df["Close"].iloc[-1]
n_anios = (df.index[-1] - df.index[0]).days / 365.25

CAGR = (final / inicio) ** (1 / n_anios) - 1
print(f"üìà CAGR: {CAGR:.2%}")
```

‚∏ª

### 4Ô∏è‚É£ Volatilidad Anualizada

Mide el riesgo o variabilidad de los rendimientos diarios, ajustado al a√±o.

F√≥rmula:

$$
\sigma_{anual} = \sigma_{diario} \cdot \sqrt{252}
$$

C√°lculo en Python:
```python
vol_daily = df["Rendimiento Diario"].std()
vol_anual = vol_daily * (252 ** 0.5)

print(f"üïí Volatilidad diaria: {vol_daily:.2%}")
print(f"üìÖ Volatilidad anualizada: {vol_anual:.2%}")
```

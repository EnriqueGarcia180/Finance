# Unidad 3: Portafolios de Inversi√≥n



### üìò 3.1 ¬øQu√© es un Portafolio de Inversi√≥n?

Un portafolio de inversi√≥n (o simplemente portafolio) es un conjunto de activos financieros en los que una persona, empresa o instituci√≥n decide invertir, con el objetivo de obtener rendimientos ajustados a su nivel de riesgo, horizonte temporal y metas financieras.

‚∏ª

üß† Concepto clave

Un portafolio no es solo una inversi√≥n, sino la combinaci√≥n diversificada de varias inversiones (acciones, bonos, efectivo, criptomonedas, bienes ra√≠ces, etc.), cuidadosamente seleccionadas y ponderadas seg√∫n una estrategia.

‚∏ª

üìä Componentes t√≠picos de un portafolio

|Activo|	Ejemplos|
|---|---|
|Acciones	|AAPL, MSFT, GOOG|
|Bonos	|Bonos del Tesoro, corporativos|
|Efectivo o equivalentes	|CETES, fondos de liquidez|
|Criptomonedas	|Bitcoin, Ethereum|
|ETFs y fondos	|SPY (S&P 500 ETF), fondos indexados|
|Otros activos	|Bienes ra√≠ces, oro, commodities|


‚∏ª

### üéØ 3.2 Objetivos de un portafolio
* Diversificaci√≥n del riesgo: No depender de un solo activo.
* Rentabilidad esperada: Maximizar los retornos ajustados al riesgo.
* Balancear riesgo-retorno: Seg√∫n el perfil del inversionista.
* Cumplir metas financieras: Retiro, educaci√≥n, compra de vivienda, etc.

‚∏ª

üìà Ejemplo simple de portafolio

Sup√≥n que un inversionista tiene $100,000 distribuidos as√≠:

|Activo	|Peso (%)|	Monto invertido|
|---|---|---|
|Acciones|	60%	|\$60,000|
|Bonos	|30%	|\$30,000|
|Criptos|	10%	|\$10,000|

Este portafolio refleja una estrategia moderada, con enfoque en crecimiento, algo de protecci√≥n con bonos y una peque√±a exposici√≥n al riesgo alto (criptomonedas).

‚∏ª

üß† ¬øQu√© hace a un portafolio ‚Äúbueno‚Äù?
* Bien diversificado
* Alineado con los objetivos y el horizonte del inversionista
* Administrado y monitoreado regularmente
* Equilibrado entre rentabilidad esperada y tolerancia al riesgo

### üìò 3.3 ¬øQu√© es la Teor√≠a de Markowitz?

La Teor√≠a de Markowitz, tambi√©n conocida como Teor√≠a Moderna de Portafolio, fue propuesta por Harry Markowitz en 1952. Su idea principal es:

üëâ No pongas todos los huevos en una sola canasta.

En otras palabras: es mejor invertir en varios activos diferentes que en uno solo, porque as√≠ reduces el riesgo sin sacrificar demasiado el rendimiento.

‚∏ª

üß† ¬øPara qu√© sirve?

Sirve para construir un portafolio de inversi√≥n balanceado, que:
* Gana lo m√°s posible
* Con el menor riesgo posible
* Aprovechando la combinaci√≥n de diferentes activos (acciones, bonos, criptomonedas, etc.)

‚∏ª

üéØ ¬øC√≥mo funciona?

Markowitz dice que al combinar activos que no se comportan igual al mismo tiempo, puedes:
* Suavizar las p√©rdidas
* Aprovechar los movimientos de cada activo
* Tener un portafolio m√°s estable

‚∏ª

üìä Ejemplo sencillo

Sup√≥n que tienes $10,000. Puedes:
* Invertir todo en una sola acci√≥n (riesgoso)
* O dividirlo entre:
* Acciones de empresas tecnol√≥gicas
* Bonos del gobierno
* Un poco de criptomonedas

‚û°Ô∏è As√≠, si las acciones bajan, puede que los bonos o criptos te protejan.

‚∏ª

üìà ¬øQu√© es la frontera eficiente?

Es una forma de ver todas las combinaciones posibles de tus inversiones.
Markowitz demostr√≥ que algunas combinaciones son mejores que otras.

La frontera eficiente es la l√≠nea que muestra las mejores combinaciones: las que te dan mayor ganancia para cada nivel de riesgo.

‚∏ª

‚úÖ En resumen:

|Concepto|	Explicaci√≥n sencilla|
|---|---|
|Portafolio	|Conjunto de inversiones diferentes|
|Diversificaci√≥n|	No invertir todo en lo mismo|
|Riesgo|	Posibilidad de perder dinero|
|Rendimiento	|Cu√°nto puedes ganar|
|Frontera eficiente	|Las mejores combinaciones de riesgo y ganancia|


‚∏ª

üéì ¬øPor qu√© es importante?

Porque te ense√±a a invertir de manera inteligente y balanceada, en lugar de apostar todo a un solo activo.
Y eso es fundamental si quieres proteger tu dinero a largo plazo.

‚∏ª

### üìò 3.4 ¬øQu√© es el Ratio de Sharpe?

El Ratio de Sharpe es una forma de medir si una inversi√≥n vale la pena tomando en cuenta cu√°nto riesgo est√°s asumiendo.

üëâ Se usa para saber qu√© tan buena es una inversi√≥n comparada con otra, no solo por cu√°nto gana, sino por cu√°n estable o riesgosa es.

‚∏ª

üß† ¬øPara qu√© sirve?

Sirve para responder esta pregunta:

‚Äú¬øEstoy recibiendo suficiente ganancia por cada unidad de riesgo que estoy asumiendo?‚Äù

‚∏ª

üìä ¬øC√≥mo se interpreta?

Valor del Ratio de Sharpe	Interpretaci√≥n general
*  $>$ 1.0	Bueno (inversi√≥n eficiente)
*  $>$ 2.0	Muy bueno
*  $>$ 3.0	Excelente
*  $<$ 1.0	Bajo (poco retorno para el riesgo)


‚∏ª

üìê F√≥rmula (simplificada)

$$ \text{Sharpe} = \frac{\text{Rendimiento de la inversi√≥n} - \text{Tasa libre de riesgo}}{\text{Riesgo (desviaci√≥n est√°ndar)}} $$

‚∏ª

üß© Explicaci√≥n con palabras simples
* Rendimiento de la inversi√≥n: cu√°nto esperas ganar.
* Tasa libre de riesgo: lo que podr√≠as ganar sin riesgo (como un bono seguro).
* Riesgo (desviaci√≥n est√°ndar): cu√°nto sube y baja tu inversi√≥n (qu√© tan inestable es).

‚û°Ô∏è Cuanto m√°s alto sea el Sharpe, mejor es tu inversi√≥n en t√©rminos de ganancia por unidad de riesgo.

‚∏ª

üìà Ejemplo sencillo

Sup√≥n que tienes dos inversiones:

|Inversi√≥n|	Gana al a√±o|	Sube y baja mucho|Ratio Sharpe|
|---|---|---|---|
|A|	10%	|Muy estable|	1.5|
|B|	15%|	Muy inestable|	0.8|

Aunque B gana m√°s, A es m√°s eficiente porque tiene mejor balance entre ganancia y riesgo.

‚∏ª

‚úÖ En resumen:

Concepto	Explicaci√≥n sencilla
Ratio Sharpe	Cu√°nto ganas por cada unidad de riesgo
Ideal	Mayor a 1.0
Uso	Comparar portafolios, fondos o estrategias
Beneficio	Eval√∫a rendimiento ajustado al riesgo
___


### üßÆ 3.5 Planteamiento del Problema

#### 3.5.1 Optimizaci√≥n de Portafolio (Frontera Eficiente de Markowitz)

Problema:
Se desea construir un portafolio √≥ptimo usando tres acciones seleccionadas (por ejemplo: AAPL, MSFT y GOOG) con datos hist√≥ricos de precios. El objetivo es determinar los pesos de inversi√≥n que maximizan el rendimiento ajustado por riesgo, es decir, el Sharpe Ratio.

‚∏ª

üìò Desarrollo Te√≥rico

La teor√≠a de Markowitz busca optimizar la asignaci√≥n de activos bajo el principio de diversificaci√≥n eficiente, construyendo la llamada Frontera Eficiente.

üéØ Objetivo:

Maximizar el Sharpe Ratio:

$$ Sharpe = \frac{R_p - R_f}{\sigma_p} $$

Donde:
* $R_p$: rendimiento esperado del portafolio
* $R_f$: tasa libre de riesgo (puede considerarse 0 si se ignora)
* $\sigma_p$: volatilidad (desviaci√≥n est√°ndar) del portafolio


### üìò 3.6 ¬øQu√© es la Frontera Eficiente?

La Frontera Eficiente es un concepto clave de la Teor√≠a Moderna de Portafolio (Markowitz), y representa las mejores combinaciones posibles de inversiones en cuanto a riesgo y rendimiento.

üëâ Es el lugar donde tienes el m√°ximo rendimiento posible para un nivel de riesgo dado, o el m√≠nimo riesgo para un rendimiento esperado.

‚∏ª

üéØ ¬øQu√© significa ser ‚Äúeficiente‚Äù?

Un portafolio es eficiente si:
* Gana m√°s que cualquier otro con el mismo riesgo
* O asume menos riesgo que otro con el mismo rendimiento

Cualquier portafolio que no est√° sobre la frontera es ineficiente, porque hay opciones mejores disponibles.


üß† ¬øPor qu√© es √∫til?

Porque te ayuda a:
* Elegir las mejores combinaciones de inversi√≥n
* Tomar decisiones informadas sobre cu√°nto riesgo asumir
* Comparar portafolios y entender si podr√≠as obtener algo mejor

‚∏ª

üìä Ejemplo pr√°ctico

Sup√≥n que tienes acciones de:
* Apple
* Google
* Microsoft

Puedes combinar esas acciones en muchas proporciones.
Al graficar el riesgo y rendimiento de cada combinaci√≥n, algunos puntos ser√°n mejores que otros: esos son los que forman la frontera eficiente.

‚∏ª

‚úÖ En resumen:

|Concepto	|Explicaci√≥n f√°cil|
|---|---|
|Frontera Eficiente|	Combinaciones √≥ptimas de riesgo/retorno|
|Portafolio eficiente|	Gana m√°s sin asumir m√°s riesgo|
|Objetivo	|Elegir el mejor portafolio posible|
|Visual	|Curva en un gr√°fico riesgo vs. retorno|



```python
# @title
# ======================================
# !pip install yfinance scipy ipywidgets --quiet

import warnings
warnings.filterwarnings("ignore", category=UserWarning)

import numpy as np
import pandas as pd
import yfinance as yf
from scipy.optimize import minimize
import matplotlib.pyplot as plt

import ipywidgets as w
from IPython.display import display, clear_output

# -------------------- Utilidades --------------------

def download_prices(tickers, start, end, auto_adjust=True):
    """
    Descarga precios (Close) con yfinance y devuelve DataFrame alineado por fecha.
    Acepta lista de tickers. Normaliza columnas y elimina vac√≠os.
    """
    df = yf.download(tickers, start=start, end=end, auto_adjust=auto_adjust, progress=False)
    if isinstance(df.columns, pd.MultiIndex):
        close = df['Close'].copy()
    else:
        # Caso de un √∫nico ticker -> normalizamos a DataFrame con una columna
        close = df[['Close']].copy()
        colname = tickers if isinstance(tickers, str) else tickers[0]
        close.columns = [colname]
    # Limpieza
    close = close.dropna(how='all').dropna(axis=1, how='all')
    # Elimina columnas vac√≠as si alg√∫n ticker fall√≥
    close = close.loc[:, close.notna().sum() > 0]
    return close

def compute_returns(prices: pd.DataFrame, kind: str = "log"):
    """
    Rendimientos logar√≠tmicos (default) o simples.
    """
    if kind.lower() == "log":
        rets = np.log(prices / prices.shift(1)).dropna()
    else:
        rets = prices.pct_change().dropna()
    return rets

def annualize_stats(rets: pd.DataFrame, periods_per_year: int = 252):
    """
    Media y covarianza anualizadas.
    """
    mean_ann = rets.mean() * periods_per_year
    cov_ann  = rets.cov() * periods_per_year
    return mean_ann, cov_ann

def sharpe_ratio(weights, mean_ann, cov_ann, rf=0.0):
    w = np.asarray(weights)
    mu  = float(w @ mean_ann.values)
    vol = float(np.sqrt(w @ cov_ann.values @ w))
    return (mu - rf) / vol if vol > 0 else -np.inf

def neg_sharpe(weights, mean_ann, cov_ann, rf=0.0):
    return -sharpe_ratio(weights, mean_ann, cov_ann, rf)

def optimize_sharpe(
    rets: pd.DataFrame,
    rf: float = 0.0,
    periods_per_year: int = 252,
    long_only: bool = True,
    n_starts: int = 25,
    seed: int = 123
):
    """
    Maximiza Sharpe (SLSQP) con m√∫ltiples inicios aleatorios.
    - long_only=True => bounds (0,1); False => (-1,1)
    - n_starts: reinicios aleatorios para evitar √≥ptimos locales.
    """
    rng = np.random.default_rng(seed)
    mean_ann, cov_ann = annualize_stats(rets, periods_per_year)
    n = rets.shape[1]

    bounds = [(0.0, 1.0)] * n if long_only else [(-1.0, 1.0)] * n
    cons   = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1.0},) #eq: equality constraint

    # Inicios: igualitario + Dirichlet (long-only) o normales (short)
    inits = [np.ones(n) / n]
    if long_only:
        inits += [rng.dirichlet(np.ones(n)) for _ in range(max(0, n_starts - 1))]
    else:
        for _ in range(n_starts - 1):
            w0 = rng.normal(0, 0.5, size=n)
            w0 = w0 / np.sum(np.abs(w0))  # normalizaci√≥n L1
            inits.append(w0)

    best_res, best_fun = None, np.inf
    for x0 in inits:
        res = minimize(
            neg_sharpe, x0=x0,
            args=(mean_ann, cov_ann, rf),
            method='SLSQP', bounds=bounds, constraints=cons, #SLSQP: Sequential Least Squares Programming
            options={'maxiter': 500, 'ftol': 1e-12, 'disp': False}
        )
        if res.success and res.fun < best_fun:
            best_res, best_fun = res, res.fun

    if best_res is None:
        raise RuntimeError("No se encontr√≥ soluci√≥n factible en la optimizaci√≥n.")

    w_opt = pd.Series(best_res.x, index=rets.columns)
    sr_opt = -best_fun
    mu_opt = float(w_opt.values @ mean_ann.values)
    vol_opt = float(np.sqrt(w_opt.values @ cov_ann.values @ w_opt.values))

    return {
        "weights": w_opt,
        "sharpe": sr_opt,
        "mu_annual": mu_opt,
        "vol_annual": vol_opt,
        "mean_ann": mean_ann,
        "cov_ann": cov_ann
    }

def simulate_random_portfolios(mean_ann: pd.Series, cov_ann: pd.DataFrame, n: int, rf: float = 0.0, long_only: bool = True, seed: int = 42):
    """
    Genera portafolios aleatorios y devuelve matriz [volatility, return, sharpe ratio].
    """
    rng = np.random.default_rng(seed)
    m, C = mean_ann.values, cov_ann.values
    res = np.zeros((3, n))
    d = len(m)
    for i in range(n):
        if long_only:
            w = rng.dirichlet(np.ones(d))
        else:
            w = rng.normal(0, 0.5, size=d)
            # normalizar a suma de pesos = 1 (permitiendo negativos)
            s = np.sum(w)
            w = w / s if s != 0 else w / np.sum(np.abs(w))
        mu  = float(w @ m)
        vol = float(np.sqrt(w @ C @ w))
        sr  = (mu - rf) / vol if vol > 0 else np.nan
        res[0, i] = vol
        res[1, i] = mu
        res[2, i] = sr
    return res

```

```python
# @title
# -------------------- UI (ipywidgets) --------------------

OPCIONES = ["AAPL","MSFT","GOOG","NVDA","META","AMZN","TSLA","BRK-B","V","JPM",
            "HWM","IBM","SPY",
            "XOM","LLY","UNH","PG","MA","KO","PEP","COST","ORCL","CSCO","NKE","ASML","TXN","ABT" ,"BTC-USD" , "ETH-USD", "USDT-USD", "XRP-USD", "LTC-USD", "ADA-USD", "DOT-USD", "BCH-USD", "XLM-USD", "LINK-USD"]

tickers_w = w.SelectMultiple(
    options=OPCIONES,
    value=("AAPL","MSFT","GOOG"),
    description="Tickers",
    rows=8
)
custom_w = w.Text(
    value="",
    description="Extra (coma):",
    placeholder="Ej: IBM, INTC, TSM"
)
start_w   = w.DatePicker(description="Inicio", value=pd.Timestamp("2023-01-01").date())
end_w     = w.DatePicker(description="Fin",    value=pd.Timestamp("2025-08-22").date())
rf_w      = w.FloatText(description="rf anual", value=0.0, step=0.001)
retkind_w = w.Dropdown(description="Rend.",
                       options=[("Log","log"),("Simple","simple")],
                       value="log")
longonly_w = w.ToggleButtons(description="Restricci√≥n",
                             options=[("Long-only","long"),("Permitir short","short")],
                             value="long")
nstarts_w = w.IntSlider(description="Reinicios", min=5, max=100, step=5, value=25)
nsim_w    = w.IntSlider(description="# Simul.", min=1000, max=50000, step=1000, value=10000)
seed_w    = w.IntText(description="Seed", value=123)

btn   = w.Button(description="Optimizar y graficar", button_style="primary", icon="bar-chart")
out   = w.Output()

def _normalize_ticker(t):
    # Ajuste com√∫n para Berkshire (yahoo usa BRK-B)
    return t.strip().upper().replace("BRK.B", "BRK-B")

def run(_):
    with out:
        clear_output()
        # Construir lista de tickers
        base = list(tickers_w.value)
        extra = [x for x in ([_normalize_ticker(t) for t in custom_w.value.split(",")] if custom_w.value else []) if x]
        tickers = list(dict.fromkeys([_normalize_ticker(t) for t in base + extra]))  # sin duplicados, preserva orden

        if not tickers:
            print("‚ö†Ô∏è Selecciona/ingresa al menos un ticker.")
            return

        start = str(start_w.value); end = str(end_w.value)
        rf = float(rf_w.value)
        kind = retkind_w.value
        long_only = (longonly_w.value == "long")
        n_starts = int(nstarts_w.value)
        n_sims = int(nsim_w.value)
        seed = int(seed_w.value)

        print("Descargando precios‚Ä¶")
        prices = download_prices(tickers, start, end, auto_adjust=True)
        if prices.empty or prices.shape[1] == 0:
            print("‚ö†Ô∏è No se obtuvieron precios v√°lidos. Revisa tickers y fechas.")
            return

        returns = compute_returns(prices, kind=kind)
        if returns.empty:
            print("‚ö†Ô∏è No hay rendimientos suficientes tras limpieza.")
            return

        print("Optimizando Sharpe‚Ä¶")
        result = optimize_sharpe(
            returns, rf=rf, periods_per_year=252,
            long_only=long_only, n_starts=n_starts, seed=seed
        )

        w_opt = (result["weights"] * 100).round(2)
        print("\nPesos √≥ptimos (%)")
        print(w_opt.to_string())
        print(f"\nSharpe √ìptimo: {result['sharpe']:.3f}")
        print(f"Rend. Anual Esperado: {result['mu_annual']:.2%}")
        print(f"Volatilidad Anual:     {result['vol_annual']:.2%}")

        # Simulaci√≥n para graficar "frontera" aleatoria
        print("\nSimulando portafolios aleatorios‚Ä¶")
        sim = simulate_random_portfolios(
            result["mean_ann"], result["cov_ann"],
            n=n_sims, rf=rf, long_only=long_only, seed=seed
        )

        opt_volatility = result["vol_annual"]
        opt_return     = result["mu_annual"]
        sharpe_opt     = result["sharpe"]

        # Gr√°fico
        plt.figure(figsize=(10, 6))
        sc = plt.scatter(sim[0, :], sim[1, :], c=sim[2, :], cmap='viridis',
                         alpha=0.35, edgecolors='none', label='Portafolios aleatorios')
        plt.colorbar(sc, label='Sharpe Ratio')
        plt.scatter(opt_volatility, opt_return, color='red', marker='*', s=220, label='√ìptimo Sharpe')
        plt.xlabel('Volatilidad Anualizada')
        plt.ylabel('Rendimiento Esperado Anualizado')
        title_suffix = "Long-only" if long_only else "Con short"
        plt.title(f'Frontera (simulada) vs Portafolio de M√°ximo Sharpe ‚Ä¢ {title_suffix}')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()

```

## Recomendaciones y Par√°metros

1) rf anual (tasa libre de riesgo anual)
* Qu√© es: La rentabilidad anual ‚Äúlibre de riesgo‚Äù usada en el Sharpe Ratio:
$$\text{Sharpe}=\frac{\mu_p - r_f}{\sigma_p}$$
*	Qu√© poner: Un n√∫mero en tanto decimal (no en %).
*	Ejemplos:
  *	0.00 ‚Üí 0% (usado cuando no quieres ajustar por rf).
	*	0.08 ‚Üí 8% anual.
	*	0.11 ‚Üí 11% (p.ej., CETES aprox. en ciertos periodos).
	* Impacto:
	*	M√°s alto ‚Üí penaliza las carteras con baja prima sobre rf; el √≥ptimo se mueve a portfolios con mayor exceso de retorno por unidad de riesgo.
	*	0.0 es √∫til did√°cticamente para ver la estructura de la frontera sin desplazarla por rf.

‚∏ª

2) Rend. (Log vs Simple)
* Simple:
$r_t=\frac{P_t-P_{t-1}}{P_{t-1}}$
  *	Suma directa en el tiempo no es exacta (compone multiplicando).
  *	Suele ser m√°s intuitivo para reportes puntuales.
*	Logar√≠tmico (Log):
$\ell_t=\ln\!\left(\frac{P_t}{P_{t-1}}\right)$
  * 	Se suma en el tiempo y aproxima bien para cambios peque√±os.
  *	Matem√°ticamente m√°s ‚Äúlimpio‚Äù para estimar medias y covarianzas.
*	Qu√© elegir:
* 	Log para optimizaci√≥n/estimaci√≥n de media‚Äìvarianza (recomendado).
*	Simple si quieres alinear con m√©tricas que alguien m√°s ya te proporciona en formato simple (reportes, backtests externos).

‚∏ª

3) Restricci√≥n (Long-only vs Permitir short)
	Long-only: pesos $\,w_i\in[0,1]$ y $\sum w_i=1$.
*	Qu√© implica: Solo compras; no hay posiciones cortas; la cartera no se apalanca.
*	Ventajas: M√°s realista para muchos mandatos, menor rotaci√≥n y menor riesgo operativo.
*	Desventajas: Frontera potencialmente menos eficiente que con short.
*	Permitir short: pesos $\,w_i\in[-1,1]$ y $\sum w_i=1$ (o similares).
*	Qu√© implica: Puedes tomar posiciones cortas (vender prestado) y compensar con largas ‚Üí mejora te√≥rica de la frontera (m√°s combinaciones).
*	Riesgos: Apalancamiento impl√≠cito, requisitos de margen, costos y riesgo operacional.
*	Qu√© elegir:
  *	Long-only si buscas algo conservador/implementable en la pr√°ctica.
  *	Short si analizas la frontera te√≥rica completa o si tu mandato permite cortos y quieres m√°xima eficiencia.

‚∏ª

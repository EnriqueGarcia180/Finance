```python
import yfinance as yf
import pandas as pd

# Tickers de ejemplo
tickers = ['AAPL', 'MSFT', 'GOOGL', 'TSLA', 'AMZN']
# 1 year
start = "2023-01-01"
end = "2024-01-01"

# Descargar datos
data = yf.download(tickers, start=start, end=end, auto_adjust=True)['Close']

# Calcular rendimiento en el período (momentum)
returns = data.iloc[-1] / data.iloc[0] - 1
momentum_ranking = returns.sort_values(ascending=False)

print("Ranking de Momentum (rendimiento anual):")
print(momentum_ranking)
```

```python
# Ranking contrario (invertido)
contrarian_ranking = returns.sort_values(ascending=True)

print("Ranking Contrario (peor rendimiento):")
print(contrarian_ranking)
```

```python
import matplotlib.pyplot as plt

momentum_ranking.plot(kind='bar', title='Momentum Ranking')
plt.ylabel('Rendimiento')
plt.grid(True)
plt.show()

contrarian_ranking.plot(kind='bar', title='Contrarian Ranking', color='red')
plt.ylabel('Rendimiento')
plt.grid(True)
plt.show()
```

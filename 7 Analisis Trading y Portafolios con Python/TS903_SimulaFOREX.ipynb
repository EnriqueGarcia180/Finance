## üß† Marco Te√≥rico

¬øQu√© es el mercado FOREX?

El mercado de divisas o FOREX (Foreign Exchange) es un mercado global descentralizado donde se negocian las monedas. Es el m√°s grande y l√≠quido del mundo, con un volumen diario superior a $6 billones de d√≥lares.

Principales conceptos:

*	Par de divisas: ejemplo EUR/USD, donde EUR es la divisa base y USD la cotizada.
*	Pip (Percentage in Point): unidad m√≠nima de cambio, normalmente $0.0001.
*	Spread: diferencia entre el precio de compra (ask) y el de venta (bid).
*	Apalancamiento (leverage): uso de capital prestado para aumentar la exposici√≥n.
*	Stop Loss: orden autom√°tica para cerrar una operaci√≥n y limitar p√©rdidas.
*	Take Profit: orden autom√°tica para cerrar una operaci√≥n con ganancias.

‚∏ª

## üìê An√°lisis T√©cnico en FOREX

El an√°lisis t√©cnico se basa en el estudio de los precios pasados para predecir movimientos futuros.

Principales herramientas:
* Medias m√≥viles: suavizan la serie de precios.
* Exponencial (EMA): da mayor peso a los datos recientes.
* Simple (SMA): promedio aritm√©tico simple.

$$
SMA_t = \frac{1}{n} \sum_{i=0}^{n-1} P_{t-i}
$$

$$
EMA_t = \alpha \cdot P_t + (1 - \alpha) \cdot EMA_{t-1}
$$

Donde:

*	$\alpha = \frac{2}{n+1}$

*	MACD: diferencia entre EMA r√°pida y lenta.

$$
\text{MACD} = EMA_{12} - EMA_{26}
$$
	‚Ä¢	RSI (√çndice de Fuerza Relativa):

$$
RSI = 100 - \frac{100}{1 + RS}
$$

Donde:
$$	\$RS = \frac{\text{Ganancia Promedio}}{\text{P√©rdida Promedio}}$$

üìå Con esto puedes crear tu portafolio de an√°lisis de FOREX en yfinance.



`````python
# !pip install yfinance plotly --quiet

import yfinance as yf
import plotly.graph_objects as go
import time
from IPython.display import clear_output
import numpy as np
import pandas as pd

# ======================
# Par√°metros
# ======================
ticker = "EURUSD=X" #"BTC-USD"
interval = "1m"        # velas de 1 minuto (es el minimo)
lookback_minutes = 60  # √∫ltimos 60 minutos
num_actualizaciones = 100 # para el for loop cuantas veces va a estar refrescando
espera_segundos = 1   # cada cu√°nto refrescar

for i in range(num_actualizaciones):
    clear_output(wait=True)

    # --- Descargar datos ---
    df = yf.download(ticker, period="1d", interval=interval,
                     progress=False, auto_adjust=True)

    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df = df.tail(lookback_minutes).dropna()

    # --- EMA, MACD ---
    df["EMA12"] = df["Close"].ewm(span=12, adjust=False).mean()
    df["EMA26"] = df["Close"].ewm(span=26, adjust=False).mean()
    df["MACD"] = df["EMA12"] - df["EMA26"]
    df["Signal"] = df["MACD"].ewm(span=9, adjust=False).mean()

    # --- RSI ---
    delta = df["Close"].diff()
    gain = np.where(delta > 0, delta, 0)
    loss = np.where(delta < 0, -delta, 0)

    avg_gain = pd.Series(gain).rolling(window=14).mean()
    avg_loss = pd.Series(loss).rolling(window=14).mean()
    rs = avg_gain / avg_loss
    df["RSI"] = 100 - (100 / (1 + rs))

    # --- Setup Trading ---
    entrada = df["Close"].iloc[-10]   # precio de hace 10 velas
    stop_loss = entrada * 0.99        # -1%
    take_profit = entrada * 1.02      # +2%
    precio_actual = df["Close"].iloc[-1]

    # --- Evaluar operaci√≥n ---
    if precio_actual >= take_profit:
        estado = "‚úÖ Se alcanz√≥ Take Profit"
    elif precio_actual <= stop_loss:
        estado = "‚ùå Se activ√≥ Stop Loss"
    else:
        estado = "‚è≥ Operaci√≥n abierta"

    # --- Gr√°fico ---
    fig = go.Figure()

    # Velas
    fig.add_trace(go.Candlestick(
        x=df.index, open=df['Open'], high=df['High'],
        low=df['Low'], close=df['Close'], name="Velas"
    ))

    # EMA
    fig.add_trace(go.Scatter(x=df.index, y=df["EMA12"],
                             line=dict(color="orange", width=1),
                             name="EMA 12"))
    fig.add_trace(go.Scatter(x=df.index, y=df["EMA26"],
                             line=dict(color="blue", width=1),
                             name="EMA 26"))

    # Layout
    fig.update_layout(
        title=f"{ticker} - Actualizaci√≥n #{i+1} ({estado})",
        xaxis_title="Tiempo",
        yaxis_title="Precio",
        xaxis_rangeslider_visible=False,
        template="seaborn",
        height=600
    )

    fig.show()

    # --- Info operaci√≥n ---
    print(f"Ticker: {ticker}")
    print(f"Precio de entrada: {entrada:.2f}")
    print(f"Stop Loss: {stop_loss:.2f}")
    print(f"Take Profit: {take_profit:.2f}")
    print(f"Precio actual: {precio_actual:.2f}")
    print(f"Estado: {estado}")
    print(f"‚è≥ Esperando {espera_segundos} seg... (Actualizaci√≥n #{i+1})")

    time.sleep(espera_segundos)
`````

`````python
inversion_inicial = 1_000_000
compra1 = 1.169727
venta1 = 1.170549
acc_compradas = inversion_inicial / compra1
acc_vendidas = acc_compradas * venta1
rendimiento = (acc_vendidas - inversion_inicial) / inversion_inicial
print(f'Inversion inicial: ${inversion_inicial:,.2f}')
print(f'Ganancia: ${acc_vendidas:,.2f}')
print(f"Rend: {rendimiento:.2%}")
`````

```python
# Librerías
import pandas as pd
import yfinance as yf
import numpy as np
# Parámetros
TICKER = "MELI"#"IVVPESOISHRS.MX" #"XRP-USD"
PERIOD = "4Y"
INTERVAL = "1d"
```

```python
# Descarga
data = yf.download(TICKER, period=PERIOD, interval=INTERVAL,    auto_adjust=True)
# multicolumns droplevel
data.columns = data.columns.droplevel(1)
display(data.tail())
```

```python
# Poner Date y CLose en nuevo df
df = data[["Close"]].copy()
display(df.tail())
```

```python
# Guardar df a CSV
pd.DataFrame(df).to_csv(f"{TICKER}.csv", index=True)
```

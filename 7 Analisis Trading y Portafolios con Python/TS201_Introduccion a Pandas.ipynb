### ðŸ§± Estructuras principales

#### 1. `Series`: una sola columna con etiquetas (Ã­ndice)
```python
import pandas as pd

serie = pd.Series([100, 105, 110], index=["Ene", "Feb", "Mar"])
print(serie)

2. DataFrame: tabla de dos dimensiones (filas y columnas)

df = pd.DataFrame({
    'Mes': ['Ene', 'Feb', 'Mar'],
    'Precio': [100, 105, 110]
})
print(df)
```



### ðŸ“¥ Lectura de datos desde archivo CSV
```python
# Supongamos que ya subiste un archivo llamado 'acciones.csv' a Colab
df = pd.read_csv('acciones.csv')
df.head()  # Muestra las primeras 5 filas

ðŸ‘‰ TambiÃ©n puedes usar archivos desde internet:

url = "https://raw.githubusercontent.com/datasets/s-and-p-500-companies/master/data/constituents.csv"
empresas = pd.read_csv(url)
empresas.head()
```

### ðŸ“¥ Escritura de datos

```python
# âœï¸ CreaciÃ³n manual de un DataFrame

data = {
    'Empresa': ['Apple', 'Microsoft', 'Google'],
    'Precio': [175.8, 312.5, 129.4],
    'Sector': ['TecnologÃ­a', 'TecnologÃ­a', 'TecnologÃ­a']
}

df = pd.DataFrame(data)
print(df)
```


1. Guardar en CSV
```python
# Guardar en CSV
df.to_csv('empresas.csv', index=False, encoding='utf-8')
```
```python
# ExplicaciÃ³n:
# - index=False evita que se guarde el Ã­ndice como columna extra
# - encoding="utf-8" asegura compatibilidad con acentos y caracteres especiales
```


2. Guardar en Excel
```python
# Guardar en Excel
df.to_excel("empresas.xlsx", index=False)
```
```python
# Nota: Pandas usa por defecto openpyxl para escribir en Excel
# Si no lo tienes instalado: pip install openpyxl
```


3. Guardar en JSON

```python
# Guardar en JSON
df.to_json("empresas.json", orient="records", force_ascii=False)

# - orient="records" â†’ genera una lista de diccionarios (formato comÃºn de JSON)
# - force_ascii=False â†’ permite caracteres como acentos y Ã±
```



4. Guardar en formato separado por tabulaciones (CSV)
```python
df.to_csv("empresas.csv", sep="\t", index=False)
```

## ðŸ”§ ManipulaciÃ³n de datos

#### ðŸ”¹ SelecciÃ³n de columnas:
```python
df['Precio']
```
ðŸ”¹ Filtrado con condiciones:
```python
df[df['Precio'] > 150]
```
ðŸ”¹ EstadÃ­sticas rÃ¡pidas:
```python
df.describe()
df['Precio'].mean()
```
ðŸ”¹ Agregar nueva columna:
```python
df['Precio con IVA'] = df['Precio'] * 1.16
```
ðŸ”¹ Eliminar columna:
```python
df = df.drop(columns=['Precio con IVA'])
```

### ðŸ“ˆ VisualizaciÃ³n rÃ¡pida con pandas + matplotlib

```python
import matplotlib.pyplot as plt

df.plot(kind='bar', x='Empresa', y='Precio', legend=False)
plt.title("Precios por Empresa")
plt.ylabel("USD")
plt.grid(True)
plt.show()
```

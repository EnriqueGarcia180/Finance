üíª Presentaci√≥n de la pr√°ctica en clase

```python
# 1. Importar librer√≠as
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 2. Descargar precios hist√≥ricos
df = yf.download("MSFT", start="2023-01-01", end="2024-01-01")
df = df[['Close']].dropna()

# 3. Gr√°fico b√°sico
plt.figure(figsize=(10, 4))
plt.plot(df['Close'], label='MSFT - Precio de cierre')
plt.title("Precio de Microsoft (2023)")
plt.xlabel("Fecha")
plt.ylabel("Precio ($)")
plt.legend()
plt.grid(True)
plt.show()

```

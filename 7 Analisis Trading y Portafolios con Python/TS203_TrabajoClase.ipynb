```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Recopilación de datos: precios históricos de 
from datetime import date

today = date.today()
print("Today's date:", today)

df = yf.download("AAPL", start="2022-01-01", end=today, auto_adjust=True)
```

```python
df
```

```python
df.describe()
```

```python
# 2. Limpieza de datos: eliminar columnas innecesarias
df = df[['Close']].dropna()
df
```

```python

# 3. Análisis simple: calcular media móvil de 30 días
df['MA30'] = df['Close'].rolling(window=30).mean()
```

```python

# 4. Visualización
plt.figure(figsize=(10, 5))
plt.plot(df['Close'], label='Precio cierre')
plt.plot(df['MA30'], label='Media móvil 30 días', linestyle='--')
plt.title("Ejemplo de Minería de Datos Financiera")
plt.legend()
plt.show()
```

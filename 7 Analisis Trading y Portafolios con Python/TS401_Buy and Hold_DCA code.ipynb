### ✅ Actividad: Simulación Buy & Hold

Planteamiento del Problema:

Tienes un archivo de datos financieros (por ejemplo, de la acción de Apple Inc. – AAPL) en un DataFrame llamado df, el cual contiene precios históricos de la acción, incluyendo el precio de cierre diario (Close).

Supón que:
* El primer día del DataFrame (por ejemplo, '2020-01-01') decides invertir $10,000 pesos.
* La inversión consiste en comprar acciones a precio de cierre ese día.
* No haces ninguna otra compra o venta intermedia.
* Al último día del DataFrame (por ejemplo, '2023-01-01'), decides vender todas las acciones al precio de cierre de ese día.

Pregunta:

¿Cuánto dinero tendrás al final?
¿Cuál fue el rendimiento total de tu inversión?


```python
import yfinance as yf
import pandas as pd
import numpy as np

# Descargar precios históricos de Apple (AAPL)
ticker = "AAPL"
start_date = "2022-01-01"
end_date = "2025-08-01"
df = yf.download(ticker, start=start_date, end=end_date, auto_adjust=True)
df = df[["Close"]]
df.dropna(inplace=True)  # Limpieza básica NaN

# isinstance() es una función incorporada de Python que se usa para verificar
# si un objeto pertenece a una clase o tipo específico,
# o a una subclase de esa clase.
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)

df
# DataFrame o la estructura de los datos para trabajar con ella
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 7))
plt.plot(df.index, df['Close'])
plt.title(f'Precio de Cierre de {ticker}')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre (USD)')
plt.grid(True)
plt.show()
```

```python
# Monto inicial invertido
inversion_inicial = 4300

# Precio de cierre del primer y último día
precio_inicio = df['Close'].iloc[0]
precio_final = df['Close'].iloc[-1]

# Número de acciones compradas el primer día
acciones_compradas = inversion_inicial / precio_inicio

# Valor final al vender todas las acciones
valor_final = acciones_compradas * precio_final

# Rendimiento obtenido
ganancia = valor_final - inversion_inicial
rendimiento_porcentual = (ganancia / inversion_inicial) * 100

# Resultados
print(f"Inversión inicial ${inversion_inicial:,.2f}")
print(f"Precio inicial: ${precio_inicio:.2f}")
print(f"Precio final: ${precio_final:.2f}")
print(f"Acciones compradas: {acciones_compradas:.4f}")
print(f"Valor final de la inversión: ${valor_final:,.2f}")
print(f"Ganancia: ${ganancia:,.2f}")
print(f"Rendimiento total: {rendimiento_porcentual:.2f}%")
```

### 📘 Actividad: Dollar Cost Averaging (DCA)

🎯 Definición

El **Dollar Cost Averaging (DCA)** es una estrategia de inversión sistemática en la que un inversor destina una cantidad fija de dinero a intervalos regulares (mensuales, quincenales, etc.) para adquirir un activo financiero, independientemente del precio del activo en ese momento.



🧠 Principio clave

Al comprar de manera constante sin tratar de predecir el mercado, el inversor termina promediando el costo por unidad del activo a lo largo del tiempo. Esto reduce el impacto de la volatilidad y evita el riesgo de comprar todo en un punto alto del mercado.



📊 Ejemplo simple de DCA

Supón que cada mes inviertes $100 en una criptomoneda o acción. Los precios de ese activo en los siguientes 4 meses son:

|Mes	| Precio (USD) |	Inversión (USD)|	Unidades compradas |
|-|-|-|-|
|1 |	\$50 |	\$100 |	2.000 |
|2 | \$25  |	\$100	|4.000 |
|3 |	\$33.33 |	\$100 |	3.000 |
|4 | \$20  |	\$100 |	5.000 |

Resultados
* Total invertido: \$400
* Total de unidades compradas: 14.000
* Costo promedio por unidad: \$400 / 14 = \$28.57
* Si el precio final es \$35, tu inversión vale \$490, con ganancia de \$90 (22.5%)



📈 Ventajas de DCA

|Ventaja	|Descripción|
|---|---|
|🎯 Reduce el riesgo	|Disminuye el impacto de comprar en picos de mercado.|
|🧘‍♂️ Menos estrés emocional|	No requiere “adivinar” el mejor momento para invertir.|
|🔄 Automático y disciplinado	|Ideal para planes a largo plazo y ahorro mensual.|
|📉 Beneficia en mercados bajistas|	Se compran más unidades cuando los precios son bajos.|




❌ Desventajas de DCA

|Desventaja|	Descripción|
|---|---|
|📉 Menor rendimiento en mercados alcistas	|Si el mercado sube consistentemente, es mejor haber invertido todo al inicio.|
|💸 Costos de transacción	|Invertir muchas veces puede generar más comisiones.|
|🕰️ Tarda más en estar completamente invertido	|El dinero sin invertir puede perder valor si no se usa pronto.|




🤔 ¿Cuándo usar DCA?

✅ Es útil cuando:
* No puedes o no quieres invertir todo de una vez.
* El mercado es volátil o incierto.
* Estás comenzando a invertir con ingresos periódicos (ej. salario mensual).
* No tienes suficiente experiencia para hacer “market timing”.


🧮 Comparación rápida: DCA vs. Inversión Única (Lump Sum)

|Estrategia	Ideal cuando…	|Riesgo	|Simplicidad|	Rendimiento esperado|
|---|---|---|---|
|DCA|	Mercado volátil o ingreso recurrente|	Bajo	|Alta	|Moderado|
|Lump Sum	|Mercado en tendencia alcista, dinero listo|	Alto	|Media|	Alto (en promedio)|



📌 Conclusión

El DCA es una estrategia sencilla y efectiva que fomenta la disciplina y reduce el riesgo asociado al momento de entrada en el mercado. Aunque puede generar menores rendimientos en mercados alcistas, es especialmente valioso para nuevos inversores y para quienes buscan invertir con regularidad sin preocuparse por la volatilidad diaria del mercado.



```python
import yfinance as yf
import pandas as pd
import numpy as np

# Descargar precios históricos de Apple (AAPL)
ticker = "AAPL"
start_date = "2022-01-01"
end_date = "2025-08-01"

df = yf.download(ticker, start=start_date, end=end_date, auto_adjust=True)
df = df[["Close"]]
df.dropna(inplace=True)  # Limpieza básica NaN

# isinstance() es una función incorporada de Python que se usa para verificar
# si un objeto pertenece a una clase o tipo específico,
# o a una subclase de esa clase.
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)

df
# DataFrame o la estructura de los datos para trabajar con ella
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define tu aportación mensual a tu Inversión DCA
aportacion = 100

# Suponiendo que 'df' es tu DataFrame con los precios y fechas
# Asegúrate de que el índice es de tipo datetime
df.index = pd.to_datetime(df.index)

# Filtramos los últimos días de cada mes
last_days = df.resample('ME').last()

# Calculamos las aportaciones de DCA (100 USD cada mes)
last_days['Aportacion'] = aportacion  # Aportación fija de 100 USD
last_days['Acciones_Compradas'] = last_days['Aportacion'] / last_days['Close']

# Calculamos la acumulación de acciones y el valor de la inversión
last_days['Acciones_Acumuladas'] = last_days['Acciones_Compradas'].cumsum()
last_days['Valor_Inversion'] = last_days['Acciones_Acumuladas'] * last_days['Close']

# Gráfico del valor de la inversión DCA
plt.figure(figsize=(20, 8))
plt.plot(last_days.index, last_days['Valor_Inversion'], label='Valor de la Inversión (DCA)', color='green')
plt.plot(df['Close'], label='Precio de Cierre', color='blue', alpha=0.3)
plt.ylabel('Valor (USD)')
plt.xlabel('Fecha')
plt.title(f'Estrategia DCA vs Precio de {ticker}')
plt.legend()
plt.grid(True)
plt.show()

# Resumen de la estrategia DCA
inversion_total = last_days['Aportacion'].sum()
valor_final = last_days['Valor_Inversion'].iloc[-1]
rendimiento = (valor_final - inversion_total) / inversion_total * 100

print(f"\nResumen de la estrategia DCA para {ticker}:")
print(f"- Total aportado: {inversion_total:.2f} USD")
print(f"- Valor final de la inversión: {valor_final:.2f} USD")
print(f"- Ganancia: {valor_final - inversion_total}")
print(f"- Rendimiento: {rendimiento:.2f}%")
```
